```python
def paint_fence(key):
    print('Painting fence', key)
    time.sleep(2)
    print('Done painting fence', key)

fence_to_paint = 10

for k in range(1, fence_to_paint+1):
    paint_fence(k)
```

```python
#Simple Multi Threading Example
for k in range(0, fence_to_paint ):
    t = threading.Thread(target=paint_fence, args=(k,))#Target
    #print("Thread:",t)
    t.start()
```

```python
#thread.join()
#Wait until current thread until the been joined has terminated.

ls = []

def t1(n):
    for k in range(1, n+1):
        ls.append(k)
        time.sleep(0.5)

def t2(n):
    for l in range(1, n+1):
        ls.append(l)
        time.sleep(0.5)
        
x = threading.Thread(target=t1, args=(5,))
x.start()

y = threading.Thread(target=t2,args=(5,))
y.start()

x.join() #Wait for x thread to finish
y.join() #Wait for y thread to finish

time.sleep(0.5)
print(ls)
```

```python
import requests
from bs4 import BeautifulSoup
import queue
import threading
#from threading import Thread
import time

def crawler(symbol):
    url = f"https://finance.yahoo.com/quote/{symbol}"
    try:
        page = requests.get(url)
        page.raise_for_status()
    except requests.exceptions.HTTPError as e:
        print("HTTP Error:", e)
    except requests.exceptions.ConnectionError as e:
        print("Error Connecting:", e)
    except requests.exceptions.Timeout as e:
        print("Timeout Error:", e)
    except requests.exceptions.RequestException as e:
        print("Request exception: ", e)

    soup = BeautifulSoup(page.content, "lxml")

    price = soup.find_all("div")[0].find_all("div", {"class": "D(ib) Mend(20px)"})[0].text
    title = soup.find_all("div")[0].find_all("div", {"class": "W(180px) Px(16px) Pt(16px) Pb(32px)"})[0].text
    print(title.split("Analyst Report:")[0], ":", price.split('(')[0])
    return price

def current_price_fetcher(stocks):
    """
    Return the current price of the stocks in the parameter lists
    Input: Stocks symbols
    Output: A list of stocks' current prices (Integer)
    """
    threads = []
    prices = []
    que = queue.Queue() #FIFO
    
    for stock in stocks:
        #Asign a task to each thread put the returned value in a queue
        t = threading.Thread(target=lambda q, arg1: q.put(crawler(arg1)), #anonymous, single use
                             args=(que, stock)) 
        t.start()
        threads.append(t)
    #Wait for those threads to finish line and join them 
    for t in threads:
        t.join() 
    while len(prices) != len(stocks):
        #Get the data from the que and append to price list
        prices.append(que.get())
    return prices, threads

start_time = time.time()
current_price, lthread = current_price_fetcher(search_array)
elapsed_time = time.time() - start_time

print(f"elapsed_time= {elapsed_time} seconds")
```

```python
#Alternate implementation using concurrent futures

# 1. Import the library
import concurrent.futures

# 2. Define the function
def our_function(symbol):
    url = f"https://finance.yahoo.com/quote/{symbol}"
    try:
        page = requests.get(url)
        page.raise_for_status()
    except requests.exceptions.HTTPError as e:
        print("HTTP Error:", e)
    except requests.exceptions.ConnectionError as e:
        print("Error Connecting:", e)
    except requests.exceptions.Timeout as e:
        print("Timeout Error:", e)
    except requests.exceptions.RequestException as e:
        print("Request exception: ", e)

    soup = BeautifulSoup(page.content, "lxml")

    price = soup.find_all("div")[0].find_all("div", {"class": "D(ib) Mend(20px)"})[0].text
    title = soup.find_all("div")[0].find_all("div", {"class": "W(180px) Px(16px) Pt(16px) Pb(32px)"})[0].text
    print(title.split("Analyst Report:")[0], ":", price.split('(')[0])
    return price


time_taken=[]
#concurrent futures thread pool executer
for l in range(0, len(np.linspace(2, 30, 15))):
    start_time = time.time() 
    # 3. Run the multithreading (You can apply any number of max worker)
    #https://docs.python.org/3/library/concurrent.futures.html
    with concurrent.futures.ThreadPoolExecutor(max_workers=np.linspace(2, 30, 15)[l] ) as executor:
        executor.map(our_function, search_array)
    total_time = time.time() - start_time

    print("Total time:", total_time)
    time_taken.append(total_time)
#time_analyze.append(total_time)
```

```python
import matplotlib.pyplot as plt

plt.plot(np.linspace(2, 12, 6), time_taken   )
plt.grid()
plt.show()
```
